# Weather data and total number of riders through out the year

```python
weather = weatheronly("../data/raw/day.csv")
weather
```

# Correlation map between temperature, humidity, windspeed

```python
sns.heatmap(weather.corr())
```

```python
sns.pairplot(weather)
```

# Scatterplot between windspeed and number of riders

```python
sns.relplot(x='Windspeed',y='Total',data=weather,hue = 'Windspeed')
```

```python
g = sns.PairGrid(weather, vars=["Windspeed", "Total"], hue="Weather Situation")
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(weather)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=3, legend=False)
```

```python
g = sns.PairGrid(weather, vars=["Weather Situation", "Total"], hue="Weather Situation")
g.map(sns.scatterplot)
```
