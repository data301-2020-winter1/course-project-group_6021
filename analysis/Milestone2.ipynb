### Question 3 & 4

#### 3. Do registered users use the bike sharing program more on a working day or a day off?

#### 4. Do casual users use the bike sharing program more on a working day or a day off?

* First we'll take a look at the the use of bike sharing by casual and registered users on a working day:

```python
dfWorkingDay = df[df["workingday"] == 1]

sns.relplot(x='weekday',y='casual',data=dfWorkingDay)
sns.relplot(x='weekday',y='registered',data=dfWorkingDay)
```

* Next, we'll take a look at the use of bike sharing by casual and registered users on a day off (i.e., a weekend or holiday):

```python
dfDayOff = df[df["workingday"] == 0]

sns.relplot(x='weekday',y='casual',data=dfDayOff)
sns.relplot(x='weekday',y='registered',data=dfDayOff)
```

Based off of these graphs, we conclude that registered users use bike sharing programs more on the weekday, while casual users use them more on the weekend. 

### Question 5
#### Did the bike sharing program get more or less popular in 2012 compared to 2011?

* In order to check this, we summed the total amound of bike share users were in 2011 and 2012 separately.

```python
dfEleven = tpd_eleven("../data/raw/day.csv")
dfTwelve = tpd_twelve("../data/raw/day.csv")

sumEleven = dfEleven['Total'].sum(axis = 0)
sumTwelve = dfTwelve['Total'].sum(axis = 0)

print("Total bike share users in 2011: " + str(sumEleven))
print("Total bike share users in 2012: " + str(sumTwelve))

```

```python
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="Month", y="Total", data=dfEleven).set_title('2011')
```

```python
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="Month", y="Total", data=dfTwelve).set_title('2012')
```

#### There were a total of 1,234,103 bike share uses in 2011.
#### There were a total of 2,049,576 bike share uses in 2012.


#### This result gives us a total use increase of 815,473 within one year.
 * In other words,there was a 60% increase in bike share uses.
